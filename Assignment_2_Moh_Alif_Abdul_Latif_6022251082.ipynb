```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Batas Sumbu X (Temperatur 0 sampai 50)
x_suhu = np.linspace(0, 50, 500) # Menggunakan linspace agar kurva mulus

def fuzzy_dingin(x):
    y = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] <= 15:
            y[i] = 1.0
        elif 15 < x[i] <= 25:
            y[i] = (25 - x[i]) / (25 - 15)
        else:
            y[i] = 0.0
    return y

def fuzzy_sedang(x):
    y = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] <= 15 or x[i] >= 35:
            y[i] = 0.0
        elif 15 < x[i] <= 25:
            y[i] = (x[i] - 15) / (25 - 15)
        elif 25 < x[i] < 35:
            y[i] = (35 - x[i]) / (35 - 25)
    return y

def fuzzy_panas(x):
    y = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] <= 25:
            y[i] = 0.0
        elif 25 < x[i] <= 35:
            y[i] = (x[i] - 25) / (35 - 25)
        else:
            y[i] = 1.0
    return y

# Hitung nilai derajat keanggotaan
y_dingin = fuzzy_dingin(x_suhu)
y_sedang = fuzzy_sedang(x_suhu)
y_panas  = fuzzy_panas(x_suhu)
```

```python
#Visualisasi grafik
plt.figure(figsize=(10, 4))

# Menggambar kurva
plt.plot(x_suhu, y_dingin, 'r', linewidth=2.5, label='Dingin')
suhu_test = 32.5
plt.axvline(x=suhu_test, color='black', linestyle='--', linewidth=1.5)

# Menandai titik potong persis di 0% Dingin
plt.plot(suhu_test, 0, 'ko') # Titik hitam di 0

# Menambahkan teks keterangan di dekat titik potong
plt.text(suhu_test + 1, 0.1, 'Dingin (0 / 0%)', fontweight='bold', color='red')
plt.text(suhu_test + 1, -0.15, 'Suhu 32.5\u00b0', fontweight='bold')

# Merapikan tampilan grafik
plt.title('Kurva Temperatur Dingin', fontsize=14, fontweight='bold')
plt.xlabel('Temperatur (t)', fontsize=12)
plt.ylabel('Derajat (\u03bc)', fontsize=12)

# Mengatur batas tampilan angka
plt.xlim(0, 50.5)
plt.ylim(0, 1.2)
plt.xticks(np.arange(0, 51, 5)) # Menampilkan angka X kelipatan 5
plt.yticks(np.arange(-0.25, 1.1, 0.25)) # Menampilkan angka Y kelipatan 0.25 (25%)

plt.legend(loc='upper left')
plt.grid(True, linestyle=':', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
#Visualisasi grafik
plt.figure(figsize=(10, 4))

# Menggambar kurva
plt.plot(x_suhu, y_sedang, 'g', linewidth=2.5, label='Sedang')
suhu_test = 32.5
plt.axvline(x=suhu_test, color='black', linestyle='--', linewidth=1.5)

# Menandai titik potong persis di 25% Sedang
plt.plot(suhu_test, 0.25, 'ko') # Titik hitam di 0.25

# Menambahkan teks keterangan di dekat titik potong
plt.text(suhu_test + 1, 0.25, 'Sedang (0.25 / 25%)', fontweight='bold', color='green')
plt.text(suhu_test + 1, -0.15, 'Suhu 32.5\u00b0', fontweight='bold')

# Merapikan tampilan grafik
plt.title('Kurva Temperatur Sedang', fontsize=14, fontweight='bold')
plt.xlabel('Temperatur (t)', fontsize=12)
plt.ylabel('Derajat (\u03bc)', fontsize=12)

# Mengatur batas tampilan angka
plt.xlim(0, 50.5)
plt.ylim(0, 1.2)
plt.xticks(np.arange(0, 51, 5)) # Menampilkan angka X kelipatan 5
plt.yticks(np.arange(-0.25, 1.1, 0.25)) # Menampilkan angka Y kelipatan 0.25 (25%)

plt.legend(loc='upper left')
plt.grid(True, linestyle=':', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
#Visualisasi grafik
plt.figure(figsize=(10, 4))

# Menggambar kurva
plt.plot(x_suhu, y_panas, 'b', linewidth=2.5, label='Panas')
suhu_test = 32.5
plt.axvline(x=suhu_test, color='black', linestyle='--', linewidth=1.5)

# Menandai titik potong persis di 75% Panas dan 25% Sedang
plt.plot(suhu_test, 0.75, 'ko') # Titik hitam di 0.75

# Menambahkan teks keterangan di dekat titik potong
plt.text(suhu_test + 1, 0.75, 'Panas (0.75 / 75%)', fontweight='bold', color='blue')
plt.text(suhu_test + 1, -0.15, 'Suhu 32.5\u00b0', fontweight='bold')

# Merapikan tampilan grafik
plt.title('Kurva Temperatur Panas', fontsize=14, fontweight='bold')
plt.xlabel('Temperatur (t)', fontsize=12)
plt.ylabel('Derajat (\u03bc)', fontsize=12)

# Mengatur batas tampilan angka
plt.xlim(0, 50.5)
plt.ylim(0, 1.2)
plt.xticks(np.arange(0, 51, 5)) # Menampilkan angka X kelipatan 5
plt.yticks(np.arange(-0.25, 1.1, 0.25)) # Menampilkan angka Y kelipatan 0.25 (25%)

plt.legend(loc='upper left')
plt.grid(True, linestyle=':', alpha=0.7)

plt.tight_layout()
plt.show()
```
